```python
def Pattern (pattern, s):
    map_p_to_w = {}    #p = pattern, w = word
    map_w_to_p = {}

    words = s.split()

    if len(pattern) != len(words):  # it checks the length of the word if its correct or not and if they match
      return False

    for char, word in zip(pattern, words):
      if char in map_p_to_w:
        if map_p_to_w[char] != word:
          return False
        else:
          map_p_to_w[char] = word

      if word in map_w_to_p:
        if map_w_to_p[word] != char:
          return False
        else:
          map_w_to_p[word] = char

        return True

#Test Cases
print(Pattern("abba","dog cat cat dog")) #true
print(Pattern("abba","dog cat cat fish")) #false
print(Pattern("aaaa","dog cat cat dog")) #false
```
